```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
```

Load the dataset

```python
data = pd.read_csv("traffic-data.csv")
print(data.head())
```

Covering the Date column into Datetime datatype

```python
data["Date"] = pd.to_datetime(data["Date"], format="%d/%m/%Y")
print(data.info())
```

Visualizing the Daily Traffic of the website

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,10))
plt.plot(data["Date"], data["Views"])
plt.title("Daily Traffic of the website")
plt.show()
```

Determining whether the dataset is stationary or seasonal

```python
result = seasonal_decompose(data["Views"], model='multiplicative', period=30)
fig = plt.figure()
fig = result.plot()
fig.set_size_inches(15,10)
```

**SARIMA preprocessing**

****

1. Determining p,d,q values

```python
pd.plotting.autocorrelation_plot(data["Views"])
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(data["Views"], lags = 100)
```

**SARIMA model**

****

```python
p, d, q = 5, 1, 2
model=sm.tsa.statespace.SARIMAX(data['Views'],
                                order=(p, d, q),
                                seasonal_order=(p, d, q, 12))
model=model.fit()
print(model.summary())
```

**Predictions**

****

Predictions for next 50 days

```python
predictions = model.predict(len(data), len(data)+50)
print(predictions)
```

Plotting the predictions

```python
data["Views"].plot(legend=True, label="Training Data",
                   figsize=(15, 10))
predictions.plot(legend=True, label="Predictions")
```
